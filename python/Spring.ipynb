```python
from matplotlib.pyplot import *
%matplotlib inline
from math import *

N = 100000 #number of iterations
Dt = 0.0001 #time interval
k = 10 #elastic constant
m = 1 #mass of the object
x_0 = 1 #spring rest lenght

t = [0] #initial times
x = [1.5] #initial position of the object
v = [1] #initial speed of the object

a = [ -(k/m)*(x[0]-x_0) ]

for i in range (1, N):
    
    x.append(x[i-1] + v[i-1] * Dt + 0.5 * a[i-1] * Dt**2)
    a.append(-(k/m)*(x[i]-x_0))
    v.append(v[i-1] + 0.5 * (a[i-1] + a[i]) * Dt)
    t.append(t[i-1] + Dt)
             
figure()
plot (t,x)
grid()
show()
```
